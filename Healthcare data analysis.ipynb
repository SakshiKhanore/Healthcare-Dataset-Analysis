```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\Documents\Projects\DA Projects\Healthcare Data Analysis\healthcare_dataset.csv')
```

```python
df1 = df.set_index(['Medical Condition', 'Date of Admission'])
df1 = df1.sort_index()
df1
```

```python
# Cleaning the dataset by ensuring consistency the 'Name' column

df1['Name'] = df1['Name'].str.strip().str.title()
df1
```

```python
df2 = df1.groupby(['Medical Condition']).size().sort_values(ascending = False)     # Gives the total count of patients per dept.
df2 = df2.to_frame(name = 'Patient Count')
df2
```

```python
df3 = df1.groupby(['Medical Condition']).mean('Length of Stay')
df3 = df3.filter(items = ['Length of Stay'])
df3
```

```python
df4 = df3.join(df2, how = 'outer')
df4.filter(items = ['Length of Stay', 'Patient Count'])
```

```python
# The above cell shows the total count of patients admitted per department & medical condition. This grouping will help us to
# uncover information like: (a) Average Length of Stay Observed Per Department and (b) Patient Distribution By Department.


# Following are the visualizations with respect to the above mentioned insights:
df4.plot(kind = 'line', figsize = (12.5, 4), subplots = True, layout = (1,2))

# BUSINESS INSIGHT : BED PLANNING AND STAFFING
# Conclusions : Although patient volume across medical conditions is nearly uniform (~9k patients each), asthma patients account for
#               significantly higher inpatient days due to longer average length of stay, indicating need for higher resource intensity per patient.
```

```python
df5 = df.set_index(['Age', 'Medical Condition'])
age_range = df5.groupby(['Medical Condition', 'Age']).size().to_frame('No. of Patients')
age_range.filter(items = ['No. of Patients'])
```

```python
arth = age_range.loc['Arthritis']
obe = age_range.loc['Obesity']
asth = age_range.loc['Asthma']
cancer = age_range.loc['Cancer']
diab = age_range.loc['Diabetes']
hyp = age_range.loc['Hypertension']
```

```python
df6 = arth.join(asth, lsuffix = '_arth', rsuffix = '_asth').join(cancer,lsuffix = '', rsuffix = '_cancer').join(diab, rsuffix = '_diab').join(hyp, rsuffix = '_hyp').join(obe, rsuffix = '_obe')
df6
```

```python
# The following graph shows age-wise distribution of patients per department

df6.plot(kind = 'line', subplots = True, layout = (3,2), figsize = (30,20))
```

```python
df7 = df.set_index(['Medical Condition'])
bill_df = df7.groupby([ 'Medical Condition']).agg(Avg_billing_amt = ('Billing Amount', 'mean'), Total_billing_amt = ('Billing Amount', 'sum'))
pd.options.display.float_format = '₹{:,.0f}'.format
avg_bill_df = bill_df.filter(items = ['Avg_billing_amt'])
pd.options.display.float_format = '₹{:,.0f}'.format
tot_bill_df = bill_df.filter(items = ['Total_billing_amt'])
bill_df
```

```python
# The following graphs demonstrate (a) Average treatment cost per condition and (b) Total cost contribution by department.
# BUSINESS INSIGHT : COST OPTIMIZATION
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ---- Bar Plot ----
axes[0].bar(bill_df.index, bill_df['Avg_billing_amt'])
axes[0].set_title('Average Billing Amount by Medical Condition')
axes[0].set_xlabel('Medical Condition')
axes[0].set_ylabel('Total Billing Amount')
axes[0].tick_params(axis='x', rotation=30)
axes[0].grid(True, linestyle='--', alpha=0.5)

# ---- Pie Plot ----
axes[1].pie(
    bill_df['Total_billing_amt'],
    labels=bill_df.index,
    autopct='%1.1f%%',
    startangle=90
)
axes[1].set_title('Billing Contribution by Medical Condition')

plt.tight_layout()
plt.show()
```

```python
df8 = df.set_index(['Gender', 'Medical Condition'])
gender = df8.groupby(['Gender', 'Medical Condition']).size().to_frame('Count')
gender
```

```python
female = gender.loc['Female']
male = gender.loc['Male']

gender_df = female.join(male, how = 'outer', lsuffix = '_female', rsuffix = '_male')

# The following plot demonstrates the gender-wise patient distribution across different medical conditions
# Conclusion : Maximum number of female patients get admitted due to Arthritis(4686) and the least are admitted due to Asthma(4553).
#              On the other hand, maximum male patients get admitted due to Diabetes, whereas least are admitted due to Obesity.
gender_df.plot(kind = 'line', subplots = True, figsize = (15,4), layout = (1,2))
```

```python
df9 = df.set_index('Medical Condition')
df9['Month'] = pd.DatetimeIndex(df9['Date of Admission']).month
month_df = df9.groupby(['Medical Condition', 'Month']).size().to_frame('Count')
month_df
```

```python
m_arth = month_df.loc['Arthritis']
m_obe = month_df.loc['Obesity']
m_asth = month_df.loc['Asthma']
m_cancer = month_df.loc['Cancer']
m_diab = month_df.loc['Diabetes']
m_hyp = month_df.loc['Hypertension']
```

```python
df10 = m_arth.join(m_asth, lsuffix = '_arth', rsuffix = '_asth').join(m_cancer,lsuffix = '', rsuffix = '_cancer').join(m_diab, rsuffix = '_diab').join(m_hyp, rsuffix = '_hyp').join(m_obe, rsuffix = '_obe')
df10
```

```python
df10['Monthly_Patient_Count'] = df10[['Count_arth',	'Count_asth',	'Count',	'Count_diab',	'Count_hyp',	'Count_obe']].sum(axis=1)
df10
```

```python
monthly_df = df10.filter(items = ['Monthly_Patient_Count'])
monthly_df.plot(kind = 'line', figsize = (12,3), xlim = (1,12), ylabel = 'Patient Count')

# The following graph shows that patient admission rate drops during February, suddenly spikes during March and goes on gradually increasing
# from April to August, ie, from Spring through Summer.
# This insight indicates the need for more staff members and efficient bed planning during April - August, least during February and the most 
# during March.
```
